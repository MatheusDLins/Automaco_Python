```python
# importar as bibliotecas

import pyautogui # automação do mause e teclado
import time # controla o tempo do programa
import pyperclip # permite copiar e colar com o python

pyautogui.PAUSE = 2

pyautogui.alert("Iniciando programa, não use o PC/Notebook, aperte 'ok' para continuar." )


#passo 1:
#-Entrar no Google Drive

#atalho do teclado
""""
Poderiamos usar os seguintes comandos caso não exista abas abertas no momento:

pyautogui.press("win")
pyautogui.press("chrome")
pyautogui.press("enter")

"""
#abrir nova aba
pyautogui.hotkey('ctrl', 't')
link = "https://drive.google.com/drive/folders/1wRTFw0sUVBjRr4hW5U9LF7DjLixRyxym"
#copiar e colar o link
pyperclip.copy(link)
pyautogui.hotkey('ctrl', 'v')
pyautogui.press("enter")
#esperar a aba carregar
time.sleep(2)


#passo 2:
#-Fazer download da base de vendas
#caso fosse necessario dois clicks [(pyautogui.click(1074, 321, clicks=2)]
pyautogui.click(1074, 321)
pyautogui.click(1159, 157)
time.sleep(3)
pyautogui.click(931, 556)
#esperar o download...
time.sleep(10)


#passo 3:
#-Calcular os indicadores(Faturamento e quantidade de produtos)
import pandas as pd

#aqui vamos ler a tabela do excel indicada no caminho
#caso queira ler outra aba do exel basta colocar [,sheet_name=2] apos o local do arquivo.
tabela = pd.read_excel(r"C:\Users\teteu\Downloads\Vendas - Dez.xlsx")
#mostrar a tabela
display(tabela)
#aqui vamos sobar a coluna com nome 'valor final'
faturamento = tabela["Valor Final"].sum() 
qtde_produtos = tabela["Quantidade"].sum()


#passo 4:
#entrar e enviar email
#entrando
pyautogui.hotkey('ctrl', 't')
linkg = ('https://mail.google.com/')
pyperclip.copy(linkg)
pyautogui.hotkey('ctrl', 'v')
pyautogui.press("enter")
time.sleep(6)

#enviando
pyautogui.click(97, 171)
time.sleep(3)
pyautogui.write("seuemail@gmail.com")
pyautogui.press("tab")

pyautogui.press("tab")
assunto = "Relatório de Vendas"
pyperclip.copy(assunto)
pyautogui.hotkey('ctrl', 'v')
pyautogui.press("tab")

texto_email = f"""
Bom dia.

O faturamento da sua loja foi de: {faturamento:,.2f}
A quantidade de produtos foi: {qtde_produtos}

ADS
Matheus Duarte.
"""
pyperclip.copy(texto_email)
pyautogui.hotkey('ctrl', 'v')
pyautogui.hotkey('ctrl', 'enter')
```
